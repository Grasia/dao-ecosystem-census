```python
# read casts.csv as votes_df
import pandas as pd
votes_df = pd.read_csv('casts.csv')
# drop unnamed column
votes_df = votes_df.drop(votes_df.columns[0], axis=1)
votes_df.head()
```

```python
# * platform: 'aragon'
votes_df['platform'] = 'aragon'
# rename orgAddress: platform_deployment_id
votes_df = votes_df.rename(columns={'orgAddress': 'platform_deployment_id'})
# voteId: proposal_id
votes_df = votes_df.rename(columns={'voteId': 'proposal_id'})
# * id: vote_id ***here we're renaming the "cast" id to vote_id
votes_df = votes_df.rename(columns={'id': 'vote_id'})
# voter: voter
# * createdAt: date
votes_df = votes_df.rename(columns={'createdAt': 'date'})
# * supports: choice
votes_df = votes_df.rename(columns={'supports': 'choice'})

votes_df = votes_df[['platform', 'platform_deployment_id', 'proposal_id', 'vote_id', 'voter', 'date', 'choice']]
votes_df.head()
```

```python
# make date a pandas date from an epoch seconds ts
votes_df['date'] = pd.to_datetime(votes_df['date'], unit='s')
votes_df.head()
```

```python
# make choice a string column
votes_df['choice'] = votes_df['choice'].astype(str)
votes_df.head()
```

```python
# save
votes_df.to_csv('aragon_votes.csv', index=False)
```
