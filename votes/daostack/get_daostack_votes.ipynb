```python
# read votes.csv as df
import pandas as pd
df = pd.read_csv('votes.csv')
# drop unnamed
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
# * platform is 'daostack'
df['platform'] = 'daostack'
# * platform_deployment_id: dao
df = df.rename(columns={'dao': 'platform_deployment_id'})
# * proposal_id: proposal
df = df.rename(columns={'proposal': 'proposal_id'})
# * id: vote_id
df = df.rename(columns={'id': 'vote_id'})
# * voter - no chg
# * createdAt: date
df = df.rename(columns={'createdAt': 'date'})
# * outcome: choice
df = df.rename(columns={'outcome': 'choice'})
# reputation: weight
df = df.rename(columns={'reputation': 'weight'})

df = df[['platform', 'platform_deployment_id', 'proposal_id', 'vote_id', 'voter', 'date', 'choice', 'weight']]
df.head()
```

```python
# make date a datetime, seconds epoch
df['date'] = pd.to_datetime(df['date'], unit='s')
# make choice a string
df['choice'] = df['choice'].astype(str)
df.head()
```

```python
# save to csv
df.to_csv('daostack_votes.csv', index=False)
```
