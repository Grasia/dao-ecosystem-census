```python
# read in 1687523727_deepdao_daos.csv
import pandas as pd
df = pd.read_csv('1687523727_deepdao_daos.csv')
df.head()
```

```python
# rename platformTitle to platform and deepdaoOrganizationId to organizationId and deepdaoDaoId to deepdaoDeploymentId
df = df.rename(columns={
    'platformTitle': 'platform',
    'deepdaoOrganizationId': 'deepdao_organization_id',
    'deepdaoDaoId': 'deepdao_deployment_id'
})
df.head()
```

```python
# add a column platform_id
import numpy as np
df['platform_id'] = np.nan
```

```python
# perform a regex on the website for a wallet address, if found, set platform_id to the wallet address, ignoring rows that have a null website
import re
pattern = re.compile('(0x[a-fA-F0-9]{40})')
df.loc[df.website.notnull(), 'platform_id'] = df[df.website.notnull()].website.str.extract(pattern, expand=False)
# show rows with a platform_id
df[df.platform_id.notnull()]
```

## Method

### Aragon

1. update aragon urls from `mainnet` to `client`, i.e., `https://mainnet.aragon.org/#/dhedge/` to `https://client.aragon.org/#/dhedge/` and pull organization address n=23
2. search for names in the DAO Analyzer data (which sources names from somewhere...) n=7
3. find names by following deployment urls from deep dao n=3
4. googling, using keywords such as the organization's name, "DAO", "Aragon", "0x" n=6
    - the client.aragon.org do not appear on google, so we need to find an inbound link to them
    - many of these sites will link to an aragon page
    - many sites are broken, ie expired ssl certs or down hosts

n=39 of 79

### Daohaus

1. match names from DAO Analyzer data n=15
2. handle duplicate matches: n=8
    1. ignore with zero votes or members
    2. prefer newer version
    3. prefer more recently created DAO
    4. cross check with website in case of uncertainty


n=23 of 23

### Daostack

1. match names from DAO Analyyzer n=4
2. google and check websites n=0

n=4 of 6

### Snapshot

1. search name on snapshot n=18
  - tiebreaker: more recent activity / number of users

n=18 of 20

### Realms

1. search name on realms
  - tiebreaker: more recent activity / number of proposals

n=18 of 18

```python
# we want to join these, merging df_manual into df, left merge
df = df.merge(df_manual, how='left', on=['deepdao_deployment_id', 'deepdao_organization_id'], validate='1:1')
df.head()
```

```python
# combine platform_id_x and platform_id_y into platform_id
df['platform_id'] = df.platform_id_x.fillna(df.platform_id_y)

# combine platform_id_source and source into platform_id_source
df['platform_id_source'] = df.platform_id_source.fillna(df.source)

# drop platform_id_x, platform_id_y, source
df = df.drop(columns=['platform_id_x', 'platform_id_y', 'source'])
df.head()
```
