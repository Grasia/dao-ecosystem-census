```python
# download aragon/organizations.csv from https://www.kaggle.com/datasets/daviddavo/dao-analyzer

import pandas as pd
df = pd.read_csv('organizations.csv')
df.head()
```

```python
# drop unnamed column
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
# add platform=aragon
df['platform'] = 'aragon'
# rename id to platform_id
df = df.rename(columns={'id': 'platform_id'})
df = df[['platform', 'platform_id', 'name']]
df.head()
```

```python
# add website column
df['website'] = 'https://etherscan.io/address/' + df['platform_id'] + '#code'
import numpy as np
df['additional'] = np.nan
df.head()
```

### get votes ("casts") count

```python
# read casts.csv
# an aragon "cast" is a vote, while an aragon "vote" is a proposal
votes_df = pd.read_csv('casts.csv')
votes_df.head()
```

```python
# keep only voteId and orgAddress
votes_df = votes_df[['voteId', 'orgAddress']]
# group by orgAddress and count votes
votes_df = votes_df.groupby('orgAddress').count()
# rename voteId to votes
votes_df = votes_df.rename(columns={'voteId': 'votes_count'})
# flatten index
votes_df = votes_df.reset_index()
votes_df.head()
```

```python
# join votes_df with df
df = df.merge(votes_df, how='left', left_on='platform_id', right_on='orgAddress', validate='one_to_one')
df.head()
```

```python
# count nan values
df['votes_count'].isna().sum()
```

```python
# count not nans
df['votes_count'].notna().sum()
```

```python
# sanity check
# how many of the daos with names have votes?
df[df['name'].notna()]['votes_count'].notna().sum()
```

```python
# and how many daos overall have names not null
df['name'].notna().sum()
```

```python
# fill nas
df['votes_count'] = df['votes_count'].fillna(0)
# make this column an integer
df['votes_count'] = df['votes_count'].astype(int)
df
```

```python
# drop orgAddress column
df = df.drop(columns=['orgAddress'])
df.head()
```

### get proposals ("votes") count

```python
# read the votes.csv
proposal_df = pd.read_csv('votes.csv')
proposal_df.head()
```

```python
# keep id and orgAddress, rename orgAddress to platform_id, group and count
proposal_df = proposal_df[['id', 'orgAddress']]
proposal_df = proposal_df.rename(columns={'orgAddress': 'platform_id'})
proposal_df = proposal_df.groupby('platform_id').count()
# flatten
proposal_df = proposal_df.reset_index()
# rename id to proposals_count
proposal_df = proposal_df.rename(columns={'id': 'proposals_count'})
proposal_df.head()
```

```python
# join proposal_df with df
df = df.merge(proposal_df, how='left', on='platform_id', validate='one_to_one')
df.head()
```

```python
# count nans
df['proposals_count'].isna().sum()
```

```python
# count not nans
df['proposals_count'].notna().sum()
```

```python
# fill nas w 0
df['proposals_count'] = df['proposals_count'].fillna(0)
# make int
df['proposals_count'] = df['proposals_count'].astype(int)
df.head()
```

```python
# save
df.to_csv('aragon_deployments.csv', index=False)
```
