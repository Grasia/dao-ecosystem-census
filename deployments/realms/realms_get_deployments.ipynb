# IMPORTANT

**This notebook needs to be run *after* the proposals and the votes fetching is complete; it relies on them for the counts values.**

```python
import os
REALMS_NODE_DIR = '../../realms_node'
REALMS_NODE_DATA_DIR = os.path.join(REALMS_NODE_DIR, 'output_deployments_2023_8_3')
```

```python
# show all the jsons in this dir
import glob
files = glob.glob(os.path.join(REALMS_NODE_DATA_DIR, '*.json'))
files
```

```python
import pandas as pd
df = pd.read_json(files[0])
for i in range(1, len(files)):
    df = df.append(pd.read_json(files[i]))
df = df.reset_index(drop=True)
df.shape
```

```python
# expand account column
df = pd.concat([df.drop(['account'], axis=1), df['account'].apply(pd.Series)], axis=1)
df.head()
```

```python
df.owner.value_counts()
```

```python
# * `platform`
df['platform'] = 'realms'
# * `platform_id`
# make platform_id from <owner>_<pubkey>
df['platform_id'] = df['owner'] + '_' + df['pubkey']
# * `name` (may be nan)
# * `website` (may be nan)
df['website'] = None
# * `additional` (may contain website, social, etc. info, may be nan)
df['additional'] = None
# * `votes_count` (the number of unique votes in the deployment)
# df['votes_count'] = None
# * `proposals_count` (the number of proposals in the deployment)
# df['proposals_count'] = None
df = df[['platform', 'platform_id', 'name', 'website', 'additional']]
df.head()
```

```python
# make id into a uuid
import uuid

df['id_to_hash'] = df['platform'] + '_' + df['platform_id']
df['id'] = df['id_to_hash'].apply(lambda x: uuid.uuid5(uuid.NAMESPACE_OID, str(x)))
df
```

```python
# drop id_to_hash
df.drop(columns=['id_to_hash'], inplace=True)
```

```python
# id: string
# platform: string
# platform_deployment_id: string
# rename platform_id to platform_deployment_id
df.rename(columns={'platform_id': 'platform_deployment_id'}, inplace=True)
# name: string
# website: string
# additional: string
# votes_count: int64
# proposals_count: int64
df = df[['id', 'platform', 'platform_deployment_id', 'name', 'website', 'additional']]
df
```

## TKTKTKTK Proposals and Votes
The proposals csv must be generated based on up-to-date vote data as this notebook relies on the proposal and on the vote data from this file.

```python
proposals_path = os.path.join('../../002_get_proposals/realms/', 'realms_proposals.csv')
proposals_df = pd.read_csv(proposals_path)
proposals_df.head()
```

```python
# group by platform_deployment_id and add up the votes
votes_count_df = proposals_df.groupby(['platform_deployment_id']).votes_count.sum().reset_index()
# rename platform_deployment_id to platform_id
votes_count_df.rename(columns={'platform_deployment_id': 'platform_id'}, inplace=True)
votes_count_df.head()
```

```python
df = df.merge(votes_count_df, how='left', on='platform_id', validate='one_to_one')
df.head()   
```

```python
df.votes_count.describe()
```

```python
# how many df rows with non null votes_count?
df[df.votes_count.notnull()].shape
```

```python
# fill the nulls with 0, means that there are no records of votes in the sdk data
df.votes_count.fillna(0, inplace=True)
df.votes_count.describe()
```

### Proposals

```python
proposals_count_df = proposals_df.groupby('platform_deployment_id').proposal_id.count()
proposals_count_df = pd.DataFrame(proposals_count_df).reset_index()
# set column name to proposals_count
proposals_count_df.rename(columns={'proposal_id': 'proposals_count'}, inplace=True)
# platform deployment id to platform id
proposals_count_df.rename(columns={'platform_deployment_id': 'platform_id'}, inplace=True)
proposals_count_df.head()
```

```python
# merge
df = df.merge(proposals_count_df, how='left', on='platform_id', validate='one_to_one')
df.head()
```

```python
df.proposals_count.describe()
```

```python
# fill with zeros if nan, means that the count of proposals is zero
df.proposals_count.fillna(0, inplace=True)
df.proposals_count.describe()
```

```python
# make votes count and proposals count integers
df.votes_count = df.votes_count.astype(int)
df.proposals_count = df.proposals_count.astype(int)
```

```python
# save as csv
df.to_csv('realms_deployments.csv', index=False)
```
