```python
import pandas as pd
proposals_df = pd.read_csv('votes.csv')
# remove unnamed column
proposals_df = proposals_df.loc[:, ~proposals_df.columns.str.contains('^Unnamed')]
proposals_df.head()
```

```python
# rename orgAddress to platform_id
proposals_df = proposals_df.rename(columns={'orgAddress': 'platform_deployment_id'})
# add column platform with value 'aragon'
proposals_df['platform'] = 'aragon'
# rename id to proposal_id
proposals_df = proposals_df.rename(columns={'id': 'proposal_id'})
# rename creator to author
proposals_df = proposals_df.rename(columns={'originalCreator': 'author'})
# rename startDate to created_at
proposals_df = proposals_df.rename(columns={'startDate': 'date'})
# rename voteNum to votes_count
proposals_df = proposals_df.rename(columns={'voteNum': 'votes_count'})
proposals_df.head()
```

```python
proposals_df = proposals_df[['platform', 'platform_deployment_id', 'proposal_id', 'author', 'date', 'votes_count']]
proposals_df.head()
```

```python
# parse date from epoch
proposals_df['date'] = pd.to_datetime(proposals_df['date'], unit='s')
proposals_df.head()
```
