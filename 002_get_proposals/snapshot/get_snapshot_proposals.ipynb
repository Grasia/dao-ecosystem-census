```python
import pandas as pd
proposals_df = pd.read_csv('./03_snapshot_proposals/1687557600_snapshot_proposals.csv')
proposals_df.head()
```

```python
# * platform
# add platform column = snapshot
proposals_df['platform'] = 'snapshot'
# * platform_deployment_id
# rename snapshotSpaceId to platform_deployment_id
proposals_df.rename(columns={'snapshotSpaceId': 'platform_deployment_id'}, inplace=True)
# * proposal_id
# rename snapshotProposalId to proposal_id
proposals_df.rename(columns={'snapshotProposalId': 'proposal_id'}, inplace=True)
# * author
# leave as-is
# * date
# rename created to date
proposals_df.rename(columns={'created': 'date'}, inplace=True)
# * votes_count
# rename votes to votes_count
proposals_df.rename(columns={'votes': 'votes_count'}, inplace=True)
# keep only the columns we need
proposals_df = proposals_df[['platform', 'platform_deployment_id', 'proposal_id', 'author', 'date', 'votes_count']]
proposals_df.head()
```

```python
# show col types
proposals_df.dtypes
```

```python
# make date column a datetime
proposals_df['date'] = pd.to_datetime(proposals_df['date'])
proposals_df.head()
```

```python
# save to csv
proposals_df.to_csv('snapshot_proposals.csv', index=False)
```
