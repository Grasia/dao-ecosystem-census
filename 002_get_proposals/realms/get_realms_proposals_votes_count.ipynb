this workbook is here to get the count of votes on a per-proposal basis for the proposals table

```python
import pandas as pd
df = pd.read_csv('realms_proposals.csv')
df.head()
```

```python
# if the votes_count column has any non-na values, quit the notebook
if df['votes_count'].notna().any():
    print('The votes_count column has non-na values. Quitting notebook.')
    quit()
```

```python
import os
REALMS_NODE_DIR = '../../realms_node'
REALMS_NODE_DATA_DIR = os.path.join(REALMS_NODE_DIR, 'output_votes_2023_6_12')

import glob
import json
```

```python
json_files = glob.glob(os.path.join(REALMS_NODE_DATA_DIR, 'proposal_votes_*.json'))
len(json_files)
```

```python
# read all the json files like 'proposal_votes_*.json' and concat into votes_df
votes_df = pd.DataFrame()
for json_file in json_files:
    with open(json_file) as f:
        votes_df = pd.concat([votes_df, pd.DataFrame(json.load(f))])
votes_df.head()
```

```python
votes_df.shape
```

```python
# expand the account column in the df
votes_df = pd.concat([votes_df.drop(['account'], axis=1), votes_df['account'].apply(pd.Series)], axis=1)
votes_df.head()
```

```python
# group by proposal, count pubkey
votes_count_df = votes_df.groupby('proposal').agg({'pubkey': 'count'}).reset_index()
votes_count_df.head()
```

```python
# rename pubkey to votes_count
votes_count_df = votes_count_df.rename(columns={'pubkey': 'votes_count'})
# rename proposal to platform_deployment_id
votes_count_df = votes_count_df.rename(columns={'proposal': 'proposal_id'})
votes_count_df.head()
```

```python
# merge with df
df = df.merge(votes_count_df, on='proposal_id', how='left', validate='1:1', suffixes=('', '_new'))
df.head()
```

```python
# overwrite the old votes_count column with the new one
df['votes_count'] = df['votes_count_new']
df = df.drop('votes_count_new', axis=1)
df.head()
```

```python
# fill na votes_count with 0
df['votes_count'] = df['votes_count'].fillna(0)
```

```python
# make votes_count an int
df['votes_count'] = df['votes_count'].astype(int)
```

```python
# save
df.to_csv('realms_proposals.csv', index=False)
```
