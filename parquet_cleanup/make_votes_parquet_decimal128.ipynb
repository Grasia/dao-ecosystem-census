```python
# make a copy of parquets/votes.parquet that saves the weights column as a decimal128
# previously they are saved as a decimal256, which is too large; let's round

import pyarrow.parquet as pq
import pyarrow as pa

table = pq.read_table('parquets/votes.parquet')
table.schema
```

```python
table.schema[7]
```

```python
# update column 7
# downcast to decimal128
table = table.set_column(7, 'weight', table.column(7).cast(pa.decimal128(38, 4), safe=False))
```

```python
import pyarrow.parquet as pq
# save to votes_decimal128.parquet in chunks
parquet_file = pq.ParquetWriter('parquets/votes_decimal128.parquet', table.schema)
for i in range(0, table.num_rows, 100000):
    parquet_file.write_table(table.slice(i, 100000))
    print('.')
parquet_file.close()
```
