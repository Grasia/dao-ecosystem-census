```python
# open ../parquets_version3/votes_new.parquet as df
import pandas as pd
votes_df = pd.read_parquet('../parquets_version3/votes_new.parquet')
proposals_df = pd.read_parquet('../parquets_version3/proposals.parquet')
deploys_df = pd.read_parquet('../parquets_version3/deployments.parquet')
# this was run with a sample of votes, not all of them
```

```python
votes_df.head()
```

```python
votes_df.merge(proposals_df, left_on='proposal_id', right_on='id', how='left', validate='many_to_one').sort_values('author').head()
# votes <-> proposals: OK
```

```python
votes_df.merge(deploys_df, left_on='deployment_id', right_on='id', how='left', validate='many_to_one').value_counts('name')
# votes <-> deploys: OK
```

```python
proposals_df.merge(deploys_df, left_on='deployment_id', right_on='id', how='left', validate='many_to_one').value_counts('name')
# proposals <-> deploys: OK
```
