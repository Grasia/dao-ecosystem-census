here we combine the manual organizations and the deepdao organizations to create a resulting table of deployments with organization ids

```python
import pandas as pd
# deepdao_deployments_by_platform_with_platform_ids.csv
# deployments_not_on_deepdao_with_manual_groups.csv
deepdao_df = pd.read_csv('deepdao_deployments_by_platform_with_platform_ids.csv')
manual_df = pd.read_csv('deployments_not_on_deepdao_with_manual_groups.csv')
# remove unnamed column from manual
manual_df = manual_df.loc[:, ~manual_df.columns.str.contains('^Unnamed')]

deepdao_df.head()
```

add proposals and votes count to deepdao df

```python
import glob

# find the *_deployments.csv files in ../DATA/deployments
deployments_csvs = glob.glob('../DATA/deployments/*_deployments.csv')

# read the deployments into a single concatenated dataframe
deployments = pd.concat([pd.read_csv(f) for f in deployments_csvs])
deployments = deployments.drop_duplicates(['platform', 'platform_id'])
deployments.head()
```

```python
# only keep platform, platfrom_id, votes_count, and proposals_count
_pre_merge_deployments = deployments[['platform', 'platform_id', 'votes_count', 'proposals_count']]
```

```python
# join the deployments with the deepdao data using platform and platform_id
deepdao_df = deepdao_df.drop_duplicates(['platform', 'platform_id'])
deepdao_df_with_deployments = pd.merge(deepdao_df, _pre_merge_deployments, how='left', left_on=['platform', 'platform_id'], right_on=['platform', 'platform_id'], validate='one_to_one')
deepdao_df_with_deployments.head()
```

```python
# we can drop the internal deployment id column since we use the platform and platform_id columns to identify deployments
manual_df = manual_df.drop(columns=['internal_deployment_id'])
```

```python
# rename the group id to manual_organization_id
manual_df = manual_df.rename(columns={'group_id': 'manual_organization_id'})
manual_df.head()
```

```python
# val countr platform_id_source
deepdao_df['deepdao_organization_id'].value_counts()
```

```python
# we want a resulting table of deployments with the following columns:
# - organization_id (may be null)
# - platform_id (not null)
# - platform (not null)
# - name
# - votes_count
# - proposals_count
# - organization_type (either manual or deepdao)
```

```python
_pre_merge_manual_df = manual_df.copy()
_pre_merge_manual_df['organization_type'] = 'manual'
# rename manual_organization_id to organization_id
_pre_merge_manual_df = _pre_merge_manual_df.rename(columns={'manual_organization_id': 'organization_id'})
# keep only the columns we want
_pre_merge_manual_df = _pre_merge_manual_df[['organization_id', 'platform_id', 'platform', 'name', 'votes_count', 'proposals_count', 'organization_type']]
_pre_merge_manual_df.head()
```

```python
_pre_merge_deepdao_df = deepdao_df.copy()
_pre_merge_deepdao_df['organization_type'] = 'deepdao'
# rename deepdao_organization_id to organization_id
_pre_merge_deepdao_df = _pre_merge_deepdao_df.rename(columns={'deepdao_organization_id': 'organization_id'})
# keep only the columns we want
_pre_merge_deepdao_df = _pre_merge_deepdao_df[['organization_id', 'platform_id', 'platform', 'name', 'votes_count', 'proposals_count', 'organization_type']]
_pre_merge_deepdao_df.head()
```

```python
# are the cols the same?
_pre_merge_manual_df.columns == _pre_merge_deepdao_df.columns
```

```python
# concat the two dataframes
deployments_with_organizations = pd.concat([_pre_merge_manual_df, _pre_merge_deepdao_df])
deployments_with_organizations.head()
```

```python
deployments_with_organizations.shape
```

```python
# this should be closer to 21k ...
# we currently have:
# - the deployments that are on deepdao
# - the deployments that are not on deepdao but are in the manual groups
# we are MISSING the ones that are not on deepdao and are not in the manual groups!
# these are in the deployments df; we need to concat and drop duplicates
deployments.head()
```

```python
# we can do this by merging the deployments_with_organizations df with the deployments df
# left merge deployments_with_organizations -> organizations
# then clean up the dupe columns
all_deployments_with_organizations = pd.merge(deployments, deployments_with_organizations, how='left', left_on=['platform', 'platform_id'], right_on=['platform', 'platform_id'], validate='one_to_one')
all_deployments_with_organizations.head()
```

```python
# org type val counts with nulls
all_deployments_with_organizations['organization_type'].value_counts()
```

```python
# how many org types are null?
all_deployments_with_organizations[all_deployments_with_organizations['organization_type'].isnull()].shape
```

```python
# looks good so far; we'll want to add another type of organization for the ones that are not on deepdao and are not in the manual groups, let's call it 'single'
# this means that this deployment is the only deployment inside of this organization
all_deployments_with_organizations['organization_type'] = all_deployments_with_organizations['organization_type'].fillna('single')
all_deployments_with_organizations['organization_type'].value_counts()
```

```python
# val counts for name_x
all_deployments_with_organizations['name_x'].value_counts()
```

```python
# val counts for name_y
all_deployments_with_organizations['name_y'].value_counts()
```

```python
# keep name_x, fill nans with name_y
all_deployments_with_organizations['name'] = all_deployments_with_organizations['name_x'].fillna(all_deployments_with_organizations['name_y'])
# drop
all_deployments_with_organizations = all_deployments_with_organizations.drop(columns=['name_x', 'name_y'])
all_deployments_with_organizations.head()
```

```python
# keep votes_count_x, fill nans with votes_count_y
all_deployments_with_organizations['votes_count'] = all_deployments_with_organizations['votes_count_x'].fillna(all_deployments_with_organizations['votes_count_y'])
# drop
all_deployments_with_organizations = all_deployments_with_organizations.drop(columns=['votes_count_x', 'votes_count_y'])
# same with proposals
all_deployments_with_organizations['proposals_count'] = all_deployments_with_organizations['proposals_count_x'].fillna(all_deployments_with_organizations['proposals_count_y'])
# drop
all_deployments_with_organizations = all_deployments_with_organizations.drop(columns=['proposals_count_x', 'proposals_count_y'])
all_deployments_with_organizations.head()
```

```python
# make organization ids for the single deployments
# we can do this by taking the platform and platform_id and hashing them as a uuid
import uuid
def make_single_organization_id(row):
    if row['organization_type'] != 'single':
        return row['organization_id']
    return str(uuid.uuid5(uuid.NAMESPACE_DNS, row['platform'] + row['platform_id']))

all_deployments_with_organizations['organization_id'] = all_deployments_with_organizations.apply(make_single_organization_id, axis=1)
all_deployments_with_organizations.head()
```

```python
# sanity check
# group the df by organization_id and show val counts of number of rows
all_deployments_with_organizations.groupby('organization_id').size().value_counts()
# looks very good
```

```python
all_deployments_with_organizations = all_deployments_with_organizations[[
    'organization_id',
    'platform_id', 'platform',
    'name', 'votes_count', 'proposals_count',
    'website', 'additional',
    'organization_type',
]]
all_deployments_with_organizations.head()
```

```python
# sort by org id
all_deployments_with_organizations = all_deployments_with_organizations.sort_values(by=['organization_id'])
all_deployments_with_organizations.head()
```

```python
# save as all_deployments_with_organizations.csv
all_deployments_with_organizations.to_csv('all_deployments_with_organizations.csv', index=False)
```
