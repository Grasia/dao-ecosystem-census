```python
import pandas as pd

all_proposals = pd.read_json('allProposalsFlat.json')
all_proposals.head()
```

```python
# read proposalVoteCounts.json
import json
with open('proposalVoteCounts.json') as f:
    proposalVoteCounts = json.load(f)
proposalVoteCounts
```

```python
# this is organized like {id: count, id: count, ...}
# make it into a df with values id and count
proposalVoteCounts = pd.DataFrame.from_dict(proposalVoteCounts, orient='index')
proposalVoteCounts = proposalVoteCounts.reset_index()
proposalVoteCounts.columns = ['id', 'voteCount']
proposalVoteCounts
```

```python
# join tables on pubkey -> id
all_proposals = all_proposals.merge(proposalVoteCounts, how='left', left_on='pubkey', right_on='id', validate='one_to_one')
all_proposals
```

```python
# how many nans in voteCount?
all_proposals['voteCount'].isna().sum()
```

```python
# how many not-nans in voteCount?
all_proposals['voteCount'].notna().sum()
```
