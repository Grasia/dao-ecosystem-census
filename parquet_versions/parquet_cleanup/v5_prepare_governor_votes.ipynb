```python
governor_votes_df = governor_votes_df[[
    'id',
    'proposal_id',
    'deployment_id',
    'platform_vote_id',
    'voter',
    'date',
    'choice',
    'weight'
]]
governor_votes_df.head()
```

```python
# how many dupe ids are there?
governor_votes_df.duplicated(subset=['id']).sum()
```

```python
governor_votes_df.dtypes
```

```python
# make uuids strings
governor_votes_df['id'] = governor_votes_df['id'].astype(str)
governor_votes_df['proposal_id'] = governor_votes_df['proposal_id'].astype(str)
governor_votes_df['deployment_id'] = governor_votes_df['deployment_id'].astype(str)
```

```python
import pyarrow as pa
# make website type a string
# governor_votes_df['website'] = governor_votes_df['website'].astype(str)
# make additional type a string
# governor_votes_df['additional'] = governor_votes_df['additional'].astype(str)
governor_votes_table = pa.Table.from_pandas(governor_votes_df, preserve_index=False)
governor_votes_table.schema
```

```python
governor_votes_table.column_names[7]
```

```python
# make weight a decimal 128
weights_as_decimal_128 = governor_votes_table['weight'].cast(pa.decimal128(38, 4))
governor_votes_table = governor_votes_table.set_column(7, 'weight', weights_as_decimal_128)
governor_votes_table.schema
```

```python
governor_votes_table.column_names[5]
```

```python
# make date with ms precision
dates_with_ms_precision = governor_votes_table['date'].cast(pa.timestamp('us'))
governor_votes_table = governor_votes_table.set_column(5, 'date', dates_with_ms_precision)
governor_votes_table.schema
```

```python
import pyarrow.parquet as pq
from pyarrow.parquet import ParquetFile
votes_v4_pf = ParquetFile('../parquets_version4/votes.parquet')
for i, record_batch in enumerate(votes_v4_pf.iter_batches(batch_size=10000)):
    # for the first chunk of records
    table = pa.Table.from_batches([record_batch])
    if i == 0:
        # create a parquet write object giving it an output file
        pqwriter = pq.ParquetWriter('../parquets_version5/votes.parquet', table.schema)            
    pqwriter.write_table(table)

pqwriter.write_table(governor_votes_table)

# close the parquet writer
pqwriter.close()
```

```python
# read '../parquets_version4/votes.parquet' size without opening
with pq.ParquetFile('../parquets_version4/votes.parquet') as pf:
    num_rows = pf.metadata.num_rows
    num_columns = len(pf.metadata.schema.names)
    print(f'Number of rows: {num_rows}')
    print(f'Number of columns: {num_columns}')
```
