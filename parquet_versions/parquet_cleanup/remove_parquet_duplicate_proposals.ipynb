```python
import pyarrow.parquet as pq

proposals = pq.read_table('../parquets_version2/proposals.parquet')
proposals
```

```python
num_proposals = proposals.shape[0]
proposals.shape
```

```python
proposals_df = proposals.to_pandas()
proposals_df
```

```python
# check if there are rows that are complete duplicates
num_dupes = proposals_df.duplicated().sum()
num_dupes
```

```python
# remove the duplicates
proposals_df = proposals_df.drop_duplicates()
assert proposals_df.shape[0] == num_proposals - num_dupes
proposals_df
```

```python
# are there duplicate platform and platform_deployment_id, proposal_id combinations?
dupes = proposals_df.groupby(['platform', 'platform_deployment_id', 'proposal_id']).size()
dupes = dupes.reset_index()
dupes.columns = ['platform', 'platform_deployment_id', 'proposal_id', 'count']
dupes
```

```python
# how many rows where count > 1
dupes[dupes['count'] > 1].shape[0]
```

```python
# save the dataframe as a parquet file proposals.parquet
import pyarrow as pa
proposals = pa.Table.from_pandas(proposals_df)
pq.write_table(proposals, '../parquets_version2/proposals.parquet')
```
