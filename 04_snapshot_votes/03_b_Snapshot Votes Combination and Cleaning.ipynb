```python
DIR_NAME = '1687517567'
```

```python
import os
import pandas as pd
directory_list = os.listdir(DIR_NAME)
# get all files including subdirectories
directory_list = [x[0] for x in os.walk(DIR_NAME)]
directory_list = directory_list[1:]

# get all files in the subdirectories
file_list = []
for directory in directory_list:
    deep_dao_dao_id = directory.split('/')[1]
    file_list += [[deep_dao_dao_id, x] for x in os.listdir(directory)]

# file_list to df
df_file_list = pd.DataFrame(file_list, columns=['deepDaoDaoId', 'snapshotProposalId'])
df_file_list.head()
```

```python
import pandas as pd

# read all csv files in the directory and compile them into one dataframe with an additional column for the organization name, from the filename
dfs = []
for index, row in df_file_list.iterrows():
    deep_dao_dao_id = row['deepDaoDaoId']
    snapshot_proposal_id = row['snapshotProposalId']

    dfs.append(pd.read_csv(f'{DIR_NAME}/{deep_dao_dao_id}/{snapshot_proposal_id}').assign(deepDaoDaoId=deep_dao_dao_id, fileProposalId=snapshot_proposal_id))

# combine all dataframes into one
df = pd.concat(dfs)
df.shape
```

```python
# make indexes snapshotId then id
# make first three columns 'deepDaoDaoId', 'proposalId', 'id'
df = df.sort_values(['deepDaoDaoId', 'proposalId', 'id'])
df.head(20)
```

```python
# save as snapshot_platforms with the datetime as a prefix
df.to_csv(f'{DIR_NAME}_snapshot_votes.csv', index=False)
df.shape
```
