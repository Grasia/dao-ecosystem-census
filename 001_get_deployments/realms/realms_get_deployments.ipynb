# IMPORTANT

This notebook needs to be run *after* the proposals and the votes fetching is complete; it relies on them for the counts values.

```python
import os
REALMS_NODE_DIR = '../../realms_node'
REALMS_NODE_DATA_DIR = os.path.join(REALMS_NODE_DIR, 'output_2023_6_10')
```

```python
import pandas as pd
df = pd.read_json(os.path.join(REALMS_NODE_DATA_DIR, 'realms.json'))
df
```

```python
# expand account column
df = pd.concat([df.drop(['account'], axis=1), df['account'].apply(pd.Series)], axis=1)
df.head()
```

```python
# * `platform`
df['platform'] = 'realms'
# * `platform_id`
# rename pubkey to platform_id
df.rename(columns={'pubkey': 'platform_id'}, inplace=True)
# * `name` (may be nan)
# * `website` (may be nan)
df['website'] = None
# * `additional` (may contain website, social, etc. info, may be nan)
df['additional'] = None
# * `votes_count` (the number of unique votes in the deployment)
# df['votes_count'] = None
# * `proposals_count` (the number of proposals in the deployment)
# df['proposals_count'] = None
df = df[['platform', 'platform_id', 'name', 'website', 'additional']]
df.head()
```

### IMPORTANT
The proposals csv must be generated based on up-to-date vote data as this notebook relies on the proposal and on the vote data from this file.

```python
proposals_path = os.path.join('../../002_get_proposals/realms/', 'realms_proposals.csv')
proposals_df = pd.read_csv(proposals_path)
proposals_df.head()
```

```python
# group by platform_deployment_id and add up the votes
votes_count_df = proposals_df.groupby(['platform_deployment_id']).votes_count.sum().reset_index()
# rename platform_deployment_id to platform_id
votes_count_df.rename(columns={'platform_deployment_id': 'platform_id'}, inplace=True)
votes_count_df.head()
```

```python
votes_count_df.boxplot(column=['votes_count'])
```

```python
df = df.merge(votes_count_df, how='left', on='platform_id', validate='one_to_one')
df.head()   
```
