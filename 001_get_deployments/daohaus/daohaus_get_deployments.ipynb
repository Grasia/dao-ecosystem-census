```python
# open moloches csv into df
import pandas as pd
df = pd.read_csv('moloches.csv')
# drop unnamed column
df = df.drop(df.columns[0], axis=1)
df.head()
```

```python
# network val counts
df['network'].value_counts()
```

```python
# version val counts
df['version'].value_counts()
```

```python
# platform
# platform_id
# name (may be nan)
# website (may be nan)
# additional (may contain website, social, etc. info, may be nan)
# add column platform
df['platform'] = 'daohaus'
# rename id to platform_id
df = df.rename(columns={'id': 'platform_id'})
# set version col to string
df['version'] = df['version'].astype(str)
# make additional column with network and version
df['additional'] = 'network: ' + df['network'] + ' version: ' + df['version']
df.head()
```

```python
# add column website
df['website'] = None
# keep only platform, platform_id, name, website, additional
df = df[['platform', 'platform_id', 'name', 'website', 'additional']]
df
```

```python
# save as dao_deployments.csv
df.to_csv('dao_deployments.csv', index=False)
```
