```python
# open moloches csv into df
import pandas as pd
df = pd.read_csv('moloches.csv')
# drop unnamed column
df = df.drop(df.columns[0], axis=1)
df.head()
```

```python
# network val counts
df['network'].value_counts()
```

```python
# version val counts
df['version'].value_counts()
```

```python
# platform
# platform_id
# name (may be nan)
# website (may be nan)
# additional (may contain website, social, etc. info, may be nan)
# add column platform
df['platform'] = 'daohaus'
# rename id to platform_id
df = df.rename(columns={'id': 'platform_id'})
# set version col to string
df['version'] = df['version'].astype(str)
# make additional column with network and version
df['additional'] = 'network: ' + df['network'] + ' version: ' + df['version']
df.head()
```

```python
# add column website
df['website'] = None
# keep only platform, platform_id, name, website, additional
df = df[['platform', 'platform_id', 'name', 'website', 'additional']]
df
```

### handle non unique ids

```python
# platform_id val counts
df['platform_id'].value_counts()
```

```python
# find where platform_id is 0xf8484b3d9e8189d4743b0ecf35a87c58b76cee7b
df[df['platform_id'] == '0xf8484b3d9e8189d4743b0ecf35a87c58b76cee7b']
combined_additional = df[df['platform_id'] == '0xf8484b3d9e8189d4743b0ecf35a87c58b76cee7b']['additional'].values
combined_additional_string = '; '.join(combined_additional)

# replace 0xf8484b3d9e8189d4743b0ecf35a87c58b76cee7b additional with combined_additional_string
df.loc[df['platform_id'] == '0xf8484b3d9e8189d4743b0ecf35a87c58b76cee7b', 'additional'] = combined_additional_string
df[df['platform_id'] == '0xf8484b3d9e8189d4743b0ecf35a87c58b76cee7b']
```

```python
# drop the second instance of 0xf8484b3d9e8189d4743b0ecf35a87c58b76cee7b
df = df.drop(df[df['platform_id'] == '0xf8484b3d9e8189d4743b0ecf35a87c58b76cee7b'].index[1])
df[df['platform_id'] == '0xf8484b3d9e8189d4743b0ecf35a87c58b76cee7b']
```

### add votes count

```python
# read votes csv into df
votes_df = pd.read_csv('votes.csv')
votes_df.head()
```

```python
# keep molochAddress and proposalAddress
votes_df = votes_df[['molochAddress', 'proposalAddress']]
# group by molochAddress and count
votes_df = votes_df.groupby('molochAddress').count()
# flatten
votes_df = votes_df.reset_index()
# rename proposalAddress to votes
votes_df = votes_df.rename(columns={'proposalAddress': 'votes_count'})
votes_df
```

```python
# join df and votes_df
df = df.merge(votes_df, how='left', left_on='platform_id', right_on='molochAddress', validate='one_to_one')
df
```

```python
# how many have votes as nan
df['votes_count'].isna().sum()
```

```python
# how many have votes
df['votes_count'].notna().sum()
```

```python
# lots of these don't have any votes....
# fill nans with 0
df['votes_count'] = df['votes_count'].fillna(0)
# make this an int
df['votes_count'] = df['votes_count'].astype(int)
df
```

### proposals count

```python
proposals_df = pd.read_csv('proposals.csv')
proposals_df.head()
```

```python
# keep id and molochAddress
proposals_df = proposals_df[['id', 'molochAddress']]
# group by molochAddress and count
proposals_df = proposals_df.groupby('molochAddress').count()
# flatten
proposals_df = proposals_df.reset_index()
# rename id to proposals_count
proposals_df = proposals_df.rename(columns={'id': 'proposals_count'})
proposals_df
```

```python
# merge df and proposals_df
df = df.merge(proposals_df, how='left', left_on='platform_id', right_on='molochAddress', validate='one_to_one')
df
```

```python
# count of proposals_count nans
df['proposals_count'].isna().sum()
```

```python
# count not nans
df['proposals_count'].notna().sum()
```

```python
df['proposals_count'] = df['proposals_count'].fillna(0)
df
```

```python
# make column an int
df['proposals_count'] = df['proposals_count'].astype(int)
# drop molochAddress_x and molochAddress_y
df = df.drop(columns=['molochAddress_x', 'molochAddress_y'])
df
```

```python
# save as dao_deployments.csv
df.to_csv('daohaus_deployments.csv', index=False)
```
