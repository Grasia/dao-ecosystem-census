```python
# download daostack/daos.csv from https://www.kaggle.com/datasets/daviddavo/dao-analyzer

import pandas as pd
df = pd.read_csv('daos.csv')
# drop unnmaed col
df = df.drop(df.columns[0], axis=1)
# add platform = daostack
df['platform'] = 'daostack'
df.head()
```

```python
# drop network and register
df = df.drop(['network', 'register'], axis=1)
# rename id to platform_id
df = df.rename(columns={'id': 'platform_id'})
df = df[['platform', 'platform_id', 'name']]
df.head()
```

```python
# make website and additional columns nans
import numpy as np
df['website'] = np.nan
df['additional'] = np.nan
df.head()
```

```python
# save to csv
df.to_csv('daostack_deployments.csv', index=False)
```
