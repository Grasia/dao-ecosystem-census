```python
# download daostack/daos.csv from https://www.kaggle.com/datasets/daviddavo/dao-analyzer

import pandas as pd
df = pd.read_csv('daos.csv')
# drop unnmaed col
df = df.drop(df.columns[0], axis=1)
# add platform = daostack
df['platform'] = 'daostack'
df.head()
```

```python
# drop network and register
df = df.drop(['network', 'register'], axis=1)
# rename id to platform_id
df = df.rename(columns={'id': 'platform_id'})
df = df[['platform', 'platform_id', 'name']]
df.head()
```

```python
# make website and additional columns nans
import numpy as np
df['website'] = np.nan
df['additional'] = np.nan
df.head()
```

### now lets get the no. of proposals per dao

```python
# read proposals as proposals_df
proposals_df = pd.read_csv('proposals.csv')
# keep only id and dao
proposals_df = proposals_df[['id', 'dao']]
proposals_df
```

```python
# group by dao, count
proposals_df = proposals_df.groupby('dao').count()
# flatten and rename id to proposals_count
proposals_df = proposals_df.reset_index()
proposals_df = proposals_df.rename(columns={'id': 'proposals_count'})
proposals_df.head()
```

```python
# value counts of proposals_df
proposals_df['proposals_count'].value_counts()
```

```python
# join proposals_df to df
df = df.merge(proposals_df, how='left', left_on='platform_id', right_on='dao', validate='1:1')
# drop dao
df = df.drop('dao', axis=1)
df.head()
```

```python
# how many nans in proposals_count
df['proposals_count'].isna().sum()
# fill nans with 0
df['proposals_count'] = df['proposals_count'].fillna(0)
df.head()
```

```python
# make proposals_count int
df['proposals_count'] = df['proposals_count'].astype(int)
```

```python
df
```

### add votes count

```python
# read votescsv as votes_df
votes_df = pd.read_csv('votes.csv')
# keep only voter and dao
votes_df = votes_df[['voter', 'dao']]
votes_df
```

```python
votes_df = votes_df.groupby('dao').count()
# flatten and rename voter to votes_count
votes_df = votes_df.reset_index()
votes_df = votes_df.rename(columns={'voter': 'votes_count'})
votes_df.head()
```

```python
# join votes_df to df
df = df.merge(votes_df, how='left', left_on='platform_id', right_on='dao', validate='1:1')
# drop dao
df = df.drop('dao', axis=1)
df.head()
```

```python
# how many nans in votes_count
df['votes_count'].isna().sum()
# fill nans with 0
df['votes_count'] = df['votes_count'].fillna(0)
df.head()
```

```python
# make votes_count int
df['votes_count'] = df['votes_count'].astype(int)
df.head()
```

```python
# reorder columns
df = df[['platform', 'platform_id', 'name', 'website', 'additional', 'votes_count', 'proposals_count']]
df.head()
```

```python
# save to csv
df.to_csv('daostack_deployments.csv', index=False)
```
