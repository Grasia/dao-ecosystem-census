We'll used the downloaded data from node to output a proposals csv

```python
import os
REALMS_NODE_DIR = '../../realms_node'
REALMS_NODE_DATA_DIR = os.path.join(REALMS_NODE_DIR, 'output_proposals_2023_8_3')
```

```python
# read all of the jsons in the directory, set the filename as a column
# and concat them all together
import glob
import pandas as pd
import json

jsons = glob.glob(os.path.join(REALMS_NODE_DATA_DIR, '*proposals.json'))

# read all the jsons into a flattened list
json_list = []
for json_file in jsons:
    with open(json_file, 'r') as f:
        read_json = json.load(f)
        # these files are a list of lists of dicts, so flatten them
        # if this is an empty list, skip
        if len(read_json) == 0:
            continue
        # if this is a list of dicts, flatten it
        if type(read_json[0]) == dict:
            read_json = [read_json]
        if type(read_json) != list:
            print(read_json)
        # for all objects in the list, set the filename as a key
        for obj in read_json:
            for i in obj:
                i['filename'] = os.path.basename(json_file)
                # replace _proposals.json at end with ''
                i['filename'] = i['filename'].replace('_proposals.json', '')
        json_list.extend(read_json)
```

```python
# remove empty lists from the json_list
json_list = [x for x in json_list if x != []]
# make json_list a 1d list instead of a list of lists
json_list = [item for sublist in json_list for item in sublist]
json_list
```

```python
len(json_list)
```

```python
# read into df
df = pd.DataFrame(json_list)
df
```

```python
# expand account column
df = pd.concat([df.drop(['account'], axis=1), df['account'].apply(pd.Series)], axis=1)
df
```

```python
df.columns
```

```python
# how many proposals are there?
len(df)
```

```python
# how many null votingCompletedAt are there?
len(df[df['votingCompletedAt'].isnull()])
```

```python
# keep only proposals where voting has finished
df = df[df['votingCompletedAt'].notnull()]
```

```python
df = df[['pubkey', 'filename', 'votingCompletedAt']]
```

```python
df.head()
```

```python
# remove 'realms_proposals_' from beginning of filename
df['filename'] = df['filename'].str.replace(r'^realm_proposals_', '')
df.head()
```

```python
# * platform
# add platform column = 'realms'
df['platform'] = 'realms'
# * platform_deployment_id
# rename governance to platform_deployment_id
df = df.rename(columns={'filename': 'platform_deployment_id'})
# * proposal_id
# rename pubkey to proposal_id
df = df.rename(columns={'pubkey': 'proposal_id'})
# * author
df['author'] = None # FIXME: pull this
# * date -- this should be as a pandas date, not as an epoch
# rename voting completed at to date
df = df.rename(columns={'votingCompletedAt': 'date'})
# convert date to pandas date
df['date'] = pd.to_datetime(df['date'], unit='s')
# * votes_count
# make votes_count column nulls
df['votes_count'] = None
df.head()
```

```python
# reordenar columnas: platform, platform_deployment_id, proposal_id, author, date, votes_count
df = df[['platform', 'platform_deployment_id', 'proposal_id', 'author', 'date', 'votes_count']]
```

```python
# let's save a list of all of the platform_deployment_ids to a json file to be used by the downloadVotes js script
proposal_ids = df['proposal_id'].unique().tolist()
with open('proposal_ids.json', 'w') as f:
    json.dump(proposal_ids, f)
```

```python
# guardar como realms_proposals.csv
df.to_csv('realms_proposals.csv', index=False)
```
